```python
pip install tiingo
```

```python
from tiingo import TiingoClient
config = {}
config['session'] = True
config['api_key'] = "xxxxxxxxxxxxxxxxxxxxxx"
client = TiingoClient(config)
```

```python
# Ticker examples for retail online internet: AMZN, BABA, EBAY, JD, LQDT, MELI, ETSY, BZUN, ...
ticker_metadata = client.get_ticker_metadata("AMZN")
```

```python
# Get latest prices, based on 3+ sources as JSON, sampled weekly
ticker_price = client.get_ticker_price("AMZN", frequency="weekly")
df = pd.DataFrame(ticker_price)
print(df)
```

```python
# Get 1 min prices, including the "open", "close" and "volume" columns
ticker_price = client.get_ticker_price("AMZN", frequency="1min", columns="open,close,volume")
df = pd.DataFrame(ticker_price)
print(df)
```

```python
# Get historical prices JSON format, sampled daily
historical_prices = client.get_ticker_price("AMZN",
                                            fmt='json',
                                            startDate='2025-11-01',
                                            endDate='2025-11-28',
                                            frequency='daily')
df = pd.DataFrame(historical_prices)
print(df)
```

```python
# Check what tickers are available, as well as metadata about each ticker
# including supported currency, exchange, and available start/end dates.
tickers = client.list_stock_tickers()
df = pd.DataFrame(tickers)
print(df)
```

```python
# Get definitions for fields available in the fundamentals-api, ticker is
# optional
definitions = client.get_fundamentals_definitions('AMZN')
df = pd.DataFrame(definitions)
print(df)
```
